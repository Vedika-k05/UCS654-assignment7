```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
data=pd.read_csv('D:/pred/assgn+lec/a7/Creditcard_data.csv')
print(data.shape)
print(data['Class'].value_counts())
```

```python
fraud=data[data['Class']==1]
non_fraud=data[data['Class']==0]
```

```python
nonfraud_sample=non_fraud.sample(n=len(fraud),random_state=42)
bd=pd.concat([fraud,nonfraud_sample])
bd=bd.sample(frac=1,random_state=42).reset_index(drop=True)
print(bd['Class'].value_counts())
```

```python
#s1-simple random
s1=bd.sample(n=10,random_state=1)
print(s1.shape)

#s2-systematic
s2=bd.iloc[::2]
print(s2.shape)

#s3-stratified
s3,_=train_test_split(bd,test_size=0.5,stratify=bd['Class'],random_state=1)
print(s3['Class'].value_counts())

#s4-simplerandom (seed diff)
s4=bd.sample(n=8,random_state=2)
print(s4.shape)
#s5-bootstrap
s5=bd.sample(n=10,replace=True,random_state=1)
print(s5.shape)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
def evaluate_model(sample):
    X=sample.drop('Class',axis=1)
    y=sample['Class']
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
    
    models={
        'Logistic Regression':LogisticRegression(max_iter=5000),
        'Random Forest':RandomForestClassifier(),
        'SVM':SVC(),
        'Naive Bayes':GaussianNB(),
        'KNN':KNeighborsClassifier()
    }
    res={}
    for i,model in models.items():
        model.fit(X_train,y_train)
        y_pred=model.predict(X_test)
        acc=accuracy_score(y_test,y_pred)
        res[i]=acc
    return res
```

```python
results={}
results['Simple Random']=evaluate_model(s1)
results['Systematic']=evaluate_model(s2)    
results['Stratified']=evaluate_model(s3)
results['Simple Random (Seed 2)']=evaluate_model(s4)
results['Bootstrap']=evaluate_model(s5)
print(results)
```

```python
r_table=pd.DataFrame(results)
print(r_table)
```
